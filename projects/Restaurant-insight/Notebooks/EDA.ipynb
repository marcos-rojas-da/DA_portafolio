# *Gráficos con plotly*

```python
fig = px.histogram(df, x="tips")
fig.show()
```

__*El histograma anterior muestra que las propinas más frecuentes son entre 2.3 y 2.49 dólares*__

```python
fig = px.box(df, x="restaurant_bill")
fig.show()
```

__*La caja anterior muestra que la mayoria de los clientes tienen consumo moderado (20 dólares), también hay consumos excepcionalmente altos*__

```python
fig = px.scatter(df, x="restaurant_bill", y="tips")
fig.show()
print()
df[["restaurant_bill", "tips"]].corr()
```

__*El gráfico de dispersión anterior muestra un sesgo hacia arriba a la derecha. A medida que aumentan los gastos del cliente, también tienden a aumentar las propinas. Esto sugiere que las propinas no son aleatorias, sino que están influenciadas por el monto total del consumo. *__

```python
px.box(df, x="waiter_sex", y="tips")
```

__*El grárfico de caja anterior nos indica el total de propinas dependiendo del género. En general, la media está casi a la misma altura, lo que indica que no hay mucha variación, excepto en algunos casos*__
